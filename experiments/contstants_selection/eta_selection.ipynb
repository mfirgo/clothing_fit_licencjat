```python
from utils.runs_and_configs import *
from utils.run_recreation import *
summary_fields = ["accuracy", "mean_target_probability","RMSE", "_step"]
eta_fields = ["mean_eta_small", "mean_eta_big", "variance_eta_big", "variance_eta_small"]
```

```python
runs_optimising_eta = filter_runs_by_config(get_runs(), "mean_eta_small_isconstant", False)
runs_optimising_eta = filter_runs_by_group(runs_optimising_eta, "no_training")
runs_optimising_eta = filter_runs_by_data_config(runs_optimising_eta, "dataset_type", "train-stratified", "train")
```

## All runs

```python
all_runs_df = get_runs_df(runs_optimising_eta, config_fields=["mean_eta_small_isconstant"], summary_fields=summary_fields+eta_fields)
for field in eta_fields:
    print(f"{field}, mean: {all_runs_df[field].mean()}, std: {all_runs_df[field].std()}")
```

## All converged runs

```python
converged_runs = filter_runs_by_tags(runs_optimising_eta, "model_converged")
converged_runs_df = get_runs_df(converged_runs, config_fields=["mean_eta_small_isconstant"], summary_fields=summary_fields+eta_fields)
for field in eta_fields:
    print(f"{field}, mean: {converged_runs_df[field].mean()}, std: {converged_runs_df[field].std()}")
```

```python
converged_runs_df.sort_values("mean_target_probability", ascending=False)
```

## Best converged runs

```python
best_converged_runs = filter_runs_by_summary(converged_runs, "accuracy", 0.2)
best_converged_runs_df = get_runs_df(best_converged_runs, summary_fields=summary_fields+eta_fields)
for field in eta_fields:
    print(f"{field}, mean: {best_converged_runs_df[field].mean()}, std: {best_converged_runs_df[field].std()}")
```

## Best runs

```python
best_runs = filter_runs_by_summary(runs_optimising_eta, "accuracy", 0.05)
best_runs_df = get_runs_df(best_runs, summary_fields=summary_fields+eta_fields)
for field in eta_fields:
    print(f"{field}, mean: {best_converged_runs_df[field].mean()}, std: {best_converged_runs_df[field].std()}")
```
